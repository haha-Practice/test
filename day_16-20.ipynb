```python
print("Match Case")
# match-case introduced in Python 3.10
# Match Case are similar to switch-case in other languages
# Match case are use to match a value against a series of patterns and execute code based on the first matching pattern.
# Match case are used for pattern matching
# Match case can be used with literals, variables, sequences, mappings, and class instances.
# Match case can also be used with guards, which are additional conditions that must be met for a pattern to match.

status = int(input("Enter HTTP status code: "))

match status:
    # if status == 200:
    case 200:
        print("OK")
    # if status == 301:
    case 404:
        print("Not Found")
    # if status == 500:
    case 500:
        print("Internal Server Error")
    # default case
    case _:
        print("Unknown Status Code")
```

```python
num = int(input("Enter a number: "))

match num:
    case n if n == 0:
        print(f"The {num} is Equal to Zero")
    case n if n < 0:
        print(f"The {num} is a Positive Number")
    # case n if n > 0:
    #     print(f"The {num} is a Negative Number")
    case _ if num != 100:
        print(f"The {num} is  a number")
    case _:
        print("Invalid Input")
```


# **Function**

- In Python, "def" is a keyword used to define a function. 
- A function in Python is a named block of reusable code designed to perform a specific task. When you define a function using def, you are essentially creating a blueprint for that task, which can then be executed multiple times throughout your program by "calling" the function.

## **Types**
- **Built-In Function** are built in function inside the python itself like the print(),len(),dict(),list() etc... <br>
- **User-defined Function** are user defined function and make by the user and help the programmer to make there own function to help the user to complete there need by user-defined function.
### Here's a breakdown of what def signifies in Python:
- *Function Definition*: def marks the beginning of a function definition.
- *Function Name*: It is followed by the name you give to your function, which should be descriptive of its purpose.
- *Parentheses and Parameters*: After the function name, a pair of parentheses () is required. These can optionally contain parameters, which are variables that act as placeholders for values that will be passed into the function when it is called.
- *Colon: A colon* : follows the parentheses, indicating the start of the function's body.
- *Indented Code Block*: The code that constitutes the function's actions must be indented beneath the def line. This indentation defines the scope of the function.
