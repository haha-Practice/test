```python
print("Type Casting")
# the type casting help us to convert the data type of a variable to another data type.
# there are two types of type casting.

print("Implicit type casting")#Automatically
print("implicit type casting is done by python automatically.")
# they convert the lower data type to higher data type automatically.
# to prevent data loss
# like integer to float, float to complex etc...
a = 5
b = 8.5
c = a + b
print(c)
print(type(c))
# in the above example python automatically converted the integer to float and then performed the addition.

print("Explicit type casting")#Manually
print("explicit type casting is done by the programmer manually.")
# they convert the higher data type to lower data type manually.
# like float to integer, complex to float etc...
a = 5
b = 8.5
c = a + int(b)
print(c)
print(type(c))
# in the above example we manually converted the float to integer and then performed the addition.
```
